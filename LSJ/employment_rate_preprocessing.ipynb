```python
import pandas as pd
```

```python
emp_rate=pd.read_csv('/content/졸업생의 취업 현황 (대학) 2024(원본).csv', header = 3) #csv파일 4번째 행을 컬럼으로 읽어오기
pd.set_option('display.max_columns', None) #head로 불러올 때 ...으로 생략되지 않게 하는 코드
emp_rate.head(10)
```

```python
emp_rate.drop(index=[0, 1, 2, 3], inplace=True) #0부터 3까지의 행 삭제
emp_rate = emp_rate.reset_index(drop=True) #인덱스를 다시 1부터 재정리

#남 여 구별되어 있는 컬럼 공통되는 컬럼으로 병합 & 컬럼 명 변경
emp_rate['1'] = pd.to_numeric(emp_rate['졸업자(A)'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 11'], errors='coerce') #컬럼 1과 2의 값을 더하기 위해 명시적으로 데이터를 숫자형으로 변환
emp_rate.drop(['졸업자(A)', 'Unnamed: 11'], axis=1, inplace=True)

emp_rate['2'] = pd.to_numeric(emp_rate['취업자(B)'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 13'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 14'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 15'], errors='coerce') +pd.to_numeric(emp_rate['Unnamed: 16'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 17'], errors='coerce') +pd.to_numeric(emp_rate['Unnamed: 18'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 19'], errors='coerce') +pd.to_numeric(emp_rate['Unnamed: 20'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 21'], errors='coerce') +pd.to_numeric(emp_rate['Unnamed: 22'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 23'], errors='coerce')
emp_rate.drop(['취업자(B)', 'Unnamed: 13', 'Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
               'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23'], axis=1, inplace=True)

emp_rate['3'] = pd.to_numeric(emp_rate['진학자(C)'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 25'], errors='coerce')
emp_rate.drop(['진학자(C)', 'Unnamed: 25'], axis=1, inplace=True)

emp_rate['4'] = pd.to_numeric(emp_rate['입대자(D)'])
emp_rate.drop(['입대자(D)'], axis=1, inplace=True)

emp_rate['5'] = pd.to_numeric(emp_rate['취업불가능자(E)'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 28'], errors='coerce')
emp_rate.drop(['취업불가능자(E)', 'Unnamed: 28'], axis=1, inplace=True)

emp_rate['6'] = pd.to_numeric(emp_rate['외국인유학생(F)'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 30'], errors='coerce')
emp_rate.drop(['외국인유학생(F)', 'Unnamed: 30'], axis=1, inplace=True)

emp_rate['7'] = pd.to_numeric(emp_rate['제외인정자(G)'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 32'], errors='coerce')
emp_rate.drop(['제외인정자(G)', 'Unnamed: 32'], axis=1, inplace=True)

emp_rate['8'] = pd.to_numeric(emp_rate['기타'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 34'], errors='coerce')
emp_rate.drop(['기타', 'Unnamed: 34'], axis=1, inplace=True)

emp_rate['9'] = pd.to_numeric(emp_rate['미상'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 36'], errors='coerce')
emp_rate.drop(['미상', 'Unnamed: 36'], axis=1, inplace=True)

emp_rate['10'] = pd.to_numeric(emp_rate['입학\n당시\n기취업자'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 39'], errors='coerce')
emp_rate.drop(['입학\n당시\n기취업자', 'Unnamed: 39'], axis=1, inplace=True)

emp_rate['11'] = pd.to_numeric(emp_rate['교내취업자'], errors='coerce') + pd.to_numeric(emp_rate['Unnamed: 41'], errors='coerce')
emp_rate.drop(['교내취업자', 'Unnamed: 41'], axis=1, inplace=True)

emp_rate.drop(['유지\n취업률', 'Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
               'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50', 'Unnamed: 51', 'Unnamed: 52', 'Unnamed: 53'], axis=1, inplace=True) #유지\n취업률\n(수시)는 2024년도부터 유지\n취업률 로 명칭이 바뀜

emp_rate.rename(columns={'취업률(%)\n[B/{A-(C+D+E+F+G)x100':'0'}, inplace=True)

new_order = ['기준연도', '학교종류', '설립구분', '지역', '상태', '학교명', '단과대학', '학과(전공)', '구분', '학과특성', '1','2','3','4','5','6','7','8','9','0','10','11']
emp_rate = emp_rate[new_order]

emp_rate.rename(columns={'0':'취업률(%)[B/{A-(C+D+E+F+G)x100', '1':'졸업자(A)', '2':'취업자(B)', '3':'진학자(C)', '4':'입대자(D)', '5':'취업불가능자(E)',
                         '6':'외국인유학생(F)', '7':'제외인정자(G)', '8':'기타', '9':'미상', '10':'입학 당시 기취업자','11':'교내취업자'}, inplace=True)

emp_rate.head()
```
